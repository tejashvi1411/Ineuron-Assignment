```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
%matplotlib inline
```

```python
df=pd.read_csv("ibm.csv")
```

```python
df[0:3]
```

```python
df.columns
```

## Age Of Employees

```python
sns.displot(data=df,x="Age",kde=True,color="lightgreen",aspect=10/6)
```

```python
int(df["Age"].mean())
```

Mean Age Of IBM Employee Is 36

Mean Age Distribution Of Employees Based On The Department They Work In:

```python
sns.catplot(data=df,x="Department",y="Age",kind="point",color="lightgreen",aspect=10/6)
```

```python
sns.catplot(data=df,x="Department",y="Age",kind="swarm",color="lightgreen",aspect=10/6)
```

```python
x=int(df[df["Department"]=="Sales"]["Age"].mean())
print("Mean Age Of Employee Working In Sales Dept Is",x)
```

```python
x=int(df[df["Department"]=="Research & Development"]["Age"].mean())
print("Mean Age Of Employee Working In Research & Development Dept Is",x)
```

```python
x=int(df[df["Department"]=="Human Resources"]["Age"].mean())
print("Mean Age Of Employee Working In Human Resources Dept Is",x)
```

```python
sns.catplot(data=df,x="Gender",y="Age",kind="point",color="lightgreen",aspect=10/6)
```

```python
sns.catplot(data=df,x="Gender",y="Age",kind="swarm",color="lightgreen",aspect=10/6)
```

```python
x=int(df[df["Gender"]=="Male"]["Age"].mean())
print("Mean Age Of Male Employee Is",x)
x=int(df[df["Gender"]=="Female"]["Age"].mean())
print("Mean Age Of Female Employee Is",x)
```
